## Matplotlib Recap

data used: USA and Canada Gas Prices

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### BASIC GRAPH

```python
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]

z = []

for value in y:
    z.append(value+2)

plt.figure(figsize=(5,3))

plt.plot(x,y,"r-.",label="x",color="Green", linewidth=3,marker=".", markersize=20,markeredgecolor="Blue")
plt.plot(x,z,"r--",label="2x",color="Red", linewidth=3, marker="^", markersize=15)
plt.title("First Graph", fontdict={"fontname": "Comic Sans MS", "fontsize": 56})
plt.xlabel("x values")
plt.ylabel("y values")

# third line
x3 = np.arange(0,4.5,0.5)
# plt.plot(x3, x3**2, "r", label="x^2")

plt.plot(x3[:5], x3[:5]**2, "r", label = "x^2")
plt.plot(x3[4:], x3[4:]**2, "r--", label = "x^2")

plt.xticks([0,1,2,3,4,5])
# plt.yticks([0,2,4,6,8,10,12])

plt.legend()

plt.savefig("mygraph.png")

plt.show()
```

## Bar Chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch("/")

bars[1].set_hatch("O")

bars[2].set_hatch("*")

plt.show()

# OR

patterns = ["/","O","*"]
for bar in bars:
    bar.set_hatch(patterns.pop())

plt.show()
```
